```python
comparison_index = pd.MultiIndex.from_tuples([
    ("Daytime", "PRE", "Random"),
    ("Daytime", "PRE", "Popular"),
    ("Daytime", "PRE", "ALS"),
    ("Daytime", "POST", "Random"),
    ("Daytime", "POST", "Popular"),
    ("Daytime", "POST", "ALS"),
    ("Nighttime", "PRE", "Random"),
    ("Nighttime", "PRE", "Popular"),
    ("Nighttime", "PRE", "ALS"),
    ("Nighttime", "POST", "Random"),
    ("Nighttime", "POST", "Popular"),
    ("Nighttime", "POST", "ALS"),
    ("Weekday", "PRE", "Random"),
    ("Weekday", "PRE", "Popular"),
    ("Weekday", "PRE", "ALS"),
    ("Weekday", "POST", "Random"),
    ("Weekday", "POST", "Popular"),
    ("Weekday", "POST", "ALS"),
    ("Weekend", "PRE", "Random"),
    ("Weekend", "PRE", "Popular"),
    ("Weekend", "PRE", "ALS"),
    ("Weekend", "POST", "Random"),
    ("Weekend", "POST", "Popular"),
    ("Weekend", "POST", "ALS")
])

comparison = pd.DataFrame([
    (rand["pre_daytime"] - rand["base_daytime"])/rand["base_daytime"],
    (pop["pre_daytime"] - pop["base_daytime"])/pop["base_daytime"],
    (ALS["pre_daytime"] - ALS["base_daytime"])/ALS["base_daytime"],

    (rand["post_daytime"] - rand["base_daytime"])/rand["base_daytime"],
    (pop["post_daytime"] - pop["base_daytime"])/pop["base_daytime"],
    (ALS["post_daytime"] - ALS["base_daytime"])/ALS["base_daytime"],

    (rand["pre_nighttime"] - rand["base_nighttime"])/rand["base_nighttime"],
    (pop["pre_nighttime"] - pop["base_nighttime"])/pop["base_nighttime"],
    (ALS["pre_nighttime"] - ALS["base_nighttime"])/ALS["base_nighttime"],

    (rand["post_nighttime"] - rand["base_nighttime"])/rand["base_nighttime"],
    (pop["post_nighttime"] - pop["base_nighttime"])/pop["base_nighttime"],
    (ALS["post_nighttime"] - ALS["base_nighttime"])/ALS["base_nighttime"],

    (rand["pre_weekday"] - rand["base_weekday"])/rand["base_weekday"],
    (pop["pre_weekday"] - pop["base_weekday"])/pop["base_weekday"],
    (ALS["pre_weekday"] - ALS["base_weekday"])/ALS["base_weekday"],

    (rand["post_weekday"] - rand["base_weekday"])/rand["base_weekday"],
    (pop["post_weekday"] - pop["base_weekday"])/pop["base_weekday"],
    (ALS["post_weekday"] - ALS["base_weekday"])/ALS["base_weekday"],

    (rand["pre_weekend"] - rand["base_weekend"])/rand["base_weekend"],
    (pop["pre_weekend"] - pop["base_weekend"])/pop["base_weekend"],
    (ALS["pre_weekend"] - ALS["base_weekend"])/ALS["base_weekend"],

    (rand["post_weekend"] - rand["base_weekend"])/rand["base_weekend"],
    (pop["post_weekend"] - pop["base_weekend"])/pop["base_weekend"],
    (ALS["post_weekend"] - ALS["base_weekend"])/ALS["base_weekend"],
], index=comparison_index)
comparison = comparison.round(3)
comparison
```

```python
def percent_change(row):
    for index in row.index:
        val = row[index]
        if val == 0:
            row[index] = "0.0%"
        elif val > 0:
            val = str(round(val * 100, 1))
            val = "+" + val + "%"
            row[index] = val
        else: # negative change
            val = str(round(float(str(val)[1:])*100, 1))
            val = "-" + val + "%"
            row[index] = val
    return row

latex(comparison.apply(percent_change, axis=1))
            
```
